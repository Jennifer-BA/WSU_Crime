```python
import pandas as pd
import requests

from bs4 import BeautifulSoup
```

```python
def get_wsu_crime_data(url):
    
    url = url
    page = requests.get(url)

    soup = BeautifulSoup(page.content, "html.parser")

    results = soup.findAll('table')[0]

    # Obtain every title of columns with tag <th>
    headers = []
    for i in results.find_all('th'):
        title = i.text
        headers.append(title)

    # Create a dataframe
    mydata = pd.DataFrame(columns = headers)

    # Create a for loop to fill mydata
    for j in results.find_all('tr')[1:]:
        row_data = j.find_all('td')
        row = [i.text for i in row_data]
        length = len(mydata)
        mydata.loc[length] = row

    return mydata
```

```python
url_list = ["https://police.wichita.edu/2020-crime-log-copy/", # 2021
            "https://police.wichita.edu/2020-crime-log/",
            "https://police.wichita.edu/2019-crime-log/",
            "https://police.wichita.edu/2018-crime-log/",
            "https://police.wichita.edu/2017-crime-log/",
            "https://police.wichita.edu/2016-crime-log/",
            "https://police.wichita.edu/2015-crime-log/",
            "https://police.wichita.edu/2014-crime-log/",
            "https://police.wichita.edu/2013-crime-log/",
            "https://police.wichita.edu/2012-crime-log/",
            "https://police.wichita.edu/2011-crime-log/",
            "https://police.wichita.edu/2010-crime-log/",
           ]
```

```python
all_crime = pd.DataFrame()

for url in url_list:
    crime = get_wsu_crime_data(url)
    all_crime = all_crime.append(crime)
    
print(all_crime.shape)
```
